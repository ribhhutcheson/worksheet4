## Part 2: Numeric Distributions

Plot distributions for a few numeric variables, such as:

- `total_hours`
- `avg_hourly_pay`
- `weekly_sales`

Then briefly reflect on what you see.


```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4.8))

sns.histplot(df, x="total_hours", bins=20, ax=axes[0])
axes[0].set_title("Total hours")

sns.histplot(df, x="avg_hourly_pay", bins=20, ax=axes[1])
axes[1].set_title("Average hourly pay")

sns.histplot(df, x="weekly_sales", bins=20, ax=axes[2])
axes[2].set_title("Weekly sales")

fig.tight_layout()
plt.show()
```

Write a short reflection here. For example:

## Part 3: Composition of a Categorical Variable

Pick one categorical variable, such as `store_type`, `region`, or `promotion_flag`, and show its composition with a suitable plot.


```python
order = df["store_type"].value_counts().index

fig, ax = plt.subplots(figsize=(8.4, 4.8))
sns.countplot(data=df, x="store_type", order=order, color="#5B7DB1", ax=ax)
ax.set_title("Composition of store type")
ax.set_xlabel("Store type")
ax.set_ylabel("Count")
fig.tight_layout()
fig.savefig(output_dir / "02_store_type_composition.png", dpi=200, bbox_inches="tight")
plt.show()
```

What does this composition tell you? Are some categories dominant or rare?
